```python
def normaliza_texto(serie):
    return (
        serie.str.lower()
             .str.strip()
             .str.replace("-", " ", regex=False)
             .str.replace("_", " ", regex=False)
             .str.replace(r"\s+", " ", regex=True)
             .str.strip()
    )
```

```python
# 2. Normalizar texto de cada columna de texto
df["sexo"] = normaliza_texto(df["sexo"])
df["idea_negocio"] = normaliza_texto(df["idea_negocio"])
df["tipo_de_emprendimiento"] = normaliza_texto(df["tipo_de_emprendimiento"])
df["barrio"] = normaliza_texto(df["barrio"])
df["línea_credito"] = normaliza_texto(df["línea_credito"])
```

```python
df["monto_del_credito"] = (
    df["monto_del_credito"]
    .str.replace("$", "", regex=False)
    .str.replace(",", "", regex=False)
    .str.replace(".00", "", regex=False)
    .str.strip()
)
```

```python
df['fecha_de_beneficio'] = pd.to_datetime(
    df['fecha_de_beneficio'],
    errors='coerce',
    format='mixed',
    dayfirst=True
).dt.strftime('%Y-%m-%d')
```

```python
import pandas as pd

df = pd.read_csv('../files/input/solicitudes_de_credito.csv', sep=';', index_col=0)

# Versión 1: solo lower + strip + colapsar espacios (NO reemplaza - ni _)
suave = (df['barrio'].str.lower().str.strip()
         .str.replace(r'\s+', ' ', regex=True).str.strip())

# Versión 2: lo que TÚ haces (reemplaza - y _ por espacio)
tu_version = (df['barrio'].str.lower().str.strip()
              .str.replace('-', ' ', regex=False)
              .str.replace('_', ' ', regex=False)
              .str.replace(r'\s+', ' ', regex=True).str.strip())

# Encuentra grupos donde la "suave" tiene varios valores pero la "tuya" los pega
comparacion = pd.DataFrame({'suave': suave, 'tuya': tu_version})
grupos = comparacion.groupby('tuya')['suave'].nunique()
sospechosos = grupos[grupos > 1].sort_values(ascending=False)

print('Barrios que tu normalización ESTÁ FUSIONANDO:')
print(sospechosos.head(20))
```

```python
import pandas as pd
df = pd.read_csv('../files/input/solicitudes_de_credito.csv', sep=';', index_col=0)

original = df['barrio'].str.lower().str.strip().str.replace(r'\s+', ' ', regex=True).str.strip()
tu_norm  = original.str.replace('-', ' ', regex=False).str.replace('_', ' ', regex=False).str.replace(r'\s+', ' ', regex=True).str.strip()

# Para cada valor "tuyo", muestra todos los originales que lo componen
fusiones = pd.DataFrame({'original': original, 'tu_norm': tu_norm})
grupos = fusiones.groupby('tu_norm')['original'].agg(lambda s: list(s.unique()))
multi = grupos[grupos.apply(len) > 1]

print(f'Tienes {len(multi)} barrios que vienen de fusionar varios originales')
for nombre, lista in multi.items():
    print(f'\n>>> "{nombre}":')
    for o in lista:
        print(f'    - "{o}"')
```
